```python
t = sp.symbols('t',real=True)
```

```python
x,y,z,f = sp.symbols('x y z f ',cls=sp.Function,real=True)
```

```python
x=x(t)
y=y(t)
z=z(t)
```

```python
r = sp.Matrix([x,y,z])
r
```

```python
f= f(x,y,z)
```

```python
integrand = f*sp.diff(r,t).norm()
```

```python
integrand
```

```python
integrand1 = integrand.subs([(f,2*x*y+sp.sqrt(z)),
                             (x,sp.cos(t)),
                             (z,t),
                             (y,sp.sin(t))])
```

```python
integrand1.doit().simplify().integrate((t,0,2*sp.pi)).evalf()
```
